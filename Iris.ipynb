##KNN
*  K Nearest Neighbour stores all the available cases.
*  For new data point it checks in its K-nearest neighbour and assigns the on the basis of similarity. 

```python
def eucledianDistance(train_input,test_input):
  return np.sqrt(np.sum(np.power((train_input-test_input),2)))
```

```python
def predictOutput(K,distance):
  count=[0]*len(dataset['target_names'])
  maximum=0
  predicted_output=0
  for i in range (K+1):
    count[distance[i][0]]+=1
    if(count[distance[i][0]]>maximum):
      maximum=count[distance[i][0]]
      predicted_output=distance[i][0]

  return predicted_output
```

```python
import operator
def findKNeighbour(train_input,test_input,output_train,K):
  distance=[]
  for train_input,target in zip(train_input,output_train):
    distance.append((target, eucledianDistance(train_input,test_input)))

  distance=sorted(distance,key=operator.itemgetter(1))
  predicted_output=predictOutput(K,distance)
  
  return predicted_output

  
```

```python
def accuracy(predicted,output_test):
  return np.mean(predicted==output_test)
```

```python
predicted=[]
for i in input_test:
  predicted.append(findKNeighbour(input_train,i,output_train,3))
score=accuracy(predicted,output_test)
print("Score %.3f"%score)
```
